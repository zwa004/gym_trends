```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib as plt
from datetime import datetime
import re
```

```python
week_att_df = pd.read_csv('Resources\Class Attendance 21 to 22.csv', nrows=66026)
week_att_df
```

```python
week_att_df.drop(['Textbox258','Textbox345','DivisionDisplay','DistrictDisplay','SpecialCredit','Textbox35','DivisionDisplay','DistrictDisplay','Textbox331','Textbox332','Textbox67','TotalForClass','Textbox33'], axis=1)
```

```python
pd.unique(week_att_df['ClassName'])
```

```python
clean_week_att = pd.DataFrame().assign(Class_name=week_att_df['ClassName'], Booking_date=week_att_df['BookingDateWeekDay'], Class_time=week_att_df['PeriodStart1'], Member=week_att_df['UserDisplay'], Status=week_att_df['BookingStatus1'])
clean_week_att
```

```python
#See unique values for class status
print(clean_week_att['Status'].unique())
```

```python
#Filter for classes only where bookings were completed
completed_bookings = clean_week_att.loc[clean_week_att['Status'] == 'Completed Booking ']
completed_bookings
```

```python
completed_bookings.loc[clean_week_att['Status'] == 'Completed Booking ']
completed_bookings
```

```python
#Find unique class names
completed_bookings["Class_name"].unique()
```

```python
#Replace Prime (52743) with PRIME (57332) since they are the same class
completed_bookings = completed_bookings.replace(to_replace='Prime (52743)', value='PRIME (57332)')
completed_bookings['Class_name'].unique()
```

```python
#Determine number of members who have attended each class
members_in_class = pd.DataFrame(completed_bookings.groupby(['Class_name'])['Member'].count())
members_in_class
```

```python
members_in_class_by_time = pd.DataFrame(completed_bookings.groupby(['Booking_date','Class_time','Class_name'], as_index=False)['Member'].count())
members_in_class_by_time.head()
```

```python
###Need to extract class time from Class_time column


#Remove Class_time column to get time
class_times = pd.DataFrame(members_in_class_by_time['Class_time'].str.split(expand=True))
class_times.head()
```

```python
#Rename 
class_times.columns = ['Num','AMPM','F','L','E']
class_times.head()
```

```python
#Combine to make Class Time
class_times['Class Time'] = class_times['Num'] + class_times['AMPM']
class_times.head()
```

```python
times_only = class_times['Class Time']
members_in_class_by_time_B = members_in_class_by_time.join(times_only)
members_in_class_by_time_B.head()
```

```python
#Resetting index in case that's what's keeping us from dropping the 2nd columnn
members_in_class_by_time_B.reset_index(drop=True)
members_in_class_by_time_B.head()
```

```python
finished_df = members_in_class_by_time_B[['Booking_date','Class_name','Member','Class Time']].copy()
finished_df.head()
```

```python
finished_df.dtypes
```

```python
finished_df['Class Time'] = pd.to_datetime(finished_df['Class Time'])
finished_df['Class Time'] = finished_df['Class Time'].dt.time
```

```python
filepath = Path('Resources/class_metrics.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)
finished_df.to_csv(filepath)
```
